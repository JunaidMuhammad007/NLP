# Finding Frequent and InFrequent words

```python
wordFreq = {}

for d in text["Review"]:
    for word in d.split(' '):
        if word in wordFreq.keys():
            wordFreq[word] += 1
        else:
            wordFreq[word] = 1
            
wordFreq
del(wordFreq[''])
```

```python
# values of keys
wordFreq.values()
```

```python
# Keys of wordfreq dictionary
wordFreq.keys()
```

```python
#finding minimum and maximum freq 
minFreqWord = min(wordFreq.values())
maxFreqWord = max(wordFreq.values())
maxFreqWords = []
minFreqWords = []

for key in wordFreq.keys():
    if wordFreq[key] == minFreqWord:
        minFreqWords.append(key)
    elif wordFreq[key] == maxFreqWord:
        maxFreqWords.append(key)
```

# Finding Longest and Shortest words

```python
vocab
```

```python
unwantedTokens = ['the',' ','to','a', 'in','an','and','by','other','I','A','&','-','3','(','2','5','+','.',',','«','»','\x85','\x80','1','x',
  'à',
  '0',
  '4',
  '8',
  '!',
  '/',
  'i',
  'S',
  ')',
  '\x93',
  '7',
  '\x95',
  ':',
  '\n',
  '\x96',
  's',
  'j']
minLen = len(vocab[0])
minLenWords = []
maxLen = len(vocab[0])
maxLenWords = []

for token in vocab:
    if token not in unwantedTokens:
        if len(token) > maxLen:
            maxLen = len(token)
        if len(token) < minLen:
            minLen = len(token)
            
for token in vocab:
    if token not in unwantedTokens:
        if len(token) == minLen:
            minLenWords.append(token)
        if len(token) == maxLen:
            maxLenWords.append(token)
        
minLen, minLenWords
```
